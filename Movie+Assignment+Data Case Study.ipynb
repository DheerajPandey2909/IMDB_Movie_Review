```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
movies=pd.read_csv("D:\Entertainment\Case Study\Movie+Assignment+Data.csv")
movies.head()

```

```python
movies["Gross"]=movies["Gross"]/1000000
movies["budget"]=movies["budget"]/1000000
movies.head()
```

```python
movies["Profit"]=movies["Gross"]-movies["budget"]
movies
```

```python
movies=movies.sort_values(by="Profit",ascending=False)
```

```python
df_by_genre=movies.loc[:,"CVotes10":"VotesnUS"]
df_by_genre
```

```python
df_by_genre[["genre_1","genre_2","genre_3"]]=movies[["genre_1","genre_2","genre_3"]]
df_by_genre["cnt"]=1
df_by_genre
```

```python
df_by_g1=df_by_genre.groupby("genre_1").aggregate(np.sum)
df_by_g2=df_by_genre.groupby("genre_2").aggregate(np.sum)
df_by_g3=df_by_genre.groupby("genre_3").aggregate(np.sum)
df_by_g1
```

```python
df_add=df_by_g1.add(df_by_g2,fill_value=0)
df_add1=df_add.add(df_by_g3,fill_value=0)
df_add1
```

```python
top_ten_genre=df_add1.loc[df_add1["cnt"]>10]
top_ten_genre.describe()
```

```python
top_ten_genre.iloc[:,0:-1]=top_ten_genre.iloc[:,0:-1].divide(top_ten_genre["cnt"],axis=0)
top_ten_genre
```

```python
top_ten_genre.loc[:,"VotesM":"VotesnUS"]=round(top_ten_genre.loc[:,"VotesM":"VotesnUS"],2)
top_ten_genre
```

```python
top_ten_genre[top_ten_genre.loc[:,"CVotes10":"CVotesnUS"].columns]=top_ten_genre[top_ten_genre.loc[:,"CVotes10":"CVotesnUS"].columns].astype(int)
```

```python
top_ten_genre
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=top_ten_genre.index,y=top_ten_genre["cnt"])
plt.show()
```
